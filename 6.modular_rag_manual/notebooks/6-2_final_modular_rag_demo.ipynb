## Modular RAG와 Agentic RAG 비교

| 구분 | Modular RAG | Agentic RAG |
|---|---|---|
| 핵심 | 기능을 모듈로 나눈다 | LLM이 다음 행동을 판단한다 |
| 흐름 | 개발자가 명시적으로 설계 | 모델이 도구 선택과 실행 순서 판단 |
| 예시 | 질문 분류 → 검색 → 근거 정리 → 답변 생성 → 검증 | 검색할지, 재검색할지, 다른 도구를 쓸지 모델이 판단 |
| 장점 | 제어 가능성이 높다 | 유연성이 높다 |
| 단점 | 설계자가 흐름을 직접 구성해야 한다 | 예측 가능성이 낮고 난이도가 높다 |
| 이번 과정 | 핵심 구현 대상 | 마지막 소개 주제 |
