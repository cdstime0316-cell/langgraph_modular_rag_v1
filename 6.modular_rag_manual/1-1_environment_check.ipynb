# 01. 환경 셋팅 확인

## 학습 목표

- Python 실행 환경을 확인한다.
- `.env` 파일의 API Key 로딩 여부를 확인한다.
- LangChain, LangGraph, Qdrant 관련 패키지가 설치되었는지 확인한다.

## 1. Python 버전 확인

Python 3.11~3.13 범위에서 실습하는 것을 권장한다.

```python
import sys

print("Python version:")
print(sys.version)
```

## 2. 환경 변수 확인

`.env` 파일에 `OPENAI_API_KEY`가 설정되어 있어야 한다.

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True, dotenv_path="../../.env")

api_key = os.getenv("OPENAI_API_KEY")
print("OPENAI_API_KEY 설정 여부:", bool(api_key))
print("OPENAI_MODEL:", os.getenv("OPENAI_MODEL", "gpt-4o-mini"))
print("EMBEDDING_MODEL:", os.getenv("EMBEDDING_MODEL", "text-embedding-3-small"))
```

## 3. 주요 패키지 import 확인

```python
import langchain
import langgraph
import qdrant_client
import pypdf

print("langchain:", getattr(langchain, "__version__", "installed"))
print("langgraph:", getattr(langgraph, "__version__", "installed"))
print("qdrant_client:", getattr(qdrant_client, "__version__", "installed"))
print("pypdf:", getattr(pypdf, "__version__", "installed"))
```
