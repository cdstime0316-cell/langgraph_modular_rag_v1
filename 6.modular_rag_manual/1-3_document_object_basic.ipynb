# 03. Document 객체 이해

## 학습 목표

- LangChain의 `Document` 객체를 이해한다.
- `page_content`와 `metadata`의 차이를 구분한다.
- RAG에서 문서를 어떻게 표현하는지 익힌다.

````python
from langchain_core.documents import Document
````

## 1. Document 객체 만들기

````python
doc = Document(
    page_content="민원인은 정중하고 정확하게 응대해야 한다.",
    metadata={
        "source": "공직자_민원응대_핵심_매뉴얼.pdf",
        "page": 7,
        "section": "민원응대 관련 기본원칙"
    }
)

print("본문:", doc.page_content)
print("메타데이터:", doc.metadata)
````

## 2. 여러 문서를 리스트로 구성하기

````python
docs = [
    Document(
        page_content="전화민원은 인사 후 소속 등을 밝히고 상담을 진행한다.",
        metadata={"page": 12, "section": "전화민원"}
    ),
    Document(
        page_content="특이민원은 감정적 맞대응을 피하고 증거자료를 확보한다.",
        metadata={"page": 15, "section": "특이민원"}
    ),
]

for d in docs:
    print(d.metadata, d.page_content)
````

## 3. RAG에서 Document가 중요한 이유

- `page_content`: 벡터 임베딩과 검색에 사용되는 본문이다.
- `metadata`: 검색 결과의 출처, 페이지, 섹션을 표시할 때 사용한다.
- 답변에 출처를 붙이려면 metadata 설계가 중요하다.

## 핵심 정리

```python
Document(
    page_content="검색 대상 본문",
    metadata={"source": "파일명", "page": 1, "section": "장 제목"}
)
```

- RAG의 기본 단위는 `Document`다.
- 좋은 RAG는 본문뿐 아니라 metadata도 함께 설계한다.
