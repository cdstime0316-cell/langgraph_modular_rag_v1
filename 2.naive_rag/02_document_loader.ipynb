# 02. 문서 로딩 (Document Loader)

RAG의 첫 번째 단계입니다. 다양한 형식의 문서를 LangChain `Document` 객체로 불러옵니다.

## Document 객체 구조
```python
Document(
    page_content="문서 내용",
    metadata={"source": "파일 경로", ...}
)
```

````python
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env")
````

## 1. TextLoader - 텍스트 파일 로딩

````python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/ai_basic.txt", encoding="utf-8")
docs = loader.load()

print(f"문서 수: {len(docs)}")
print(f"문서 타입: {type(docs[0])}")
print(f"메타데이터: {docs[0].metadata}")
print(f"내용 앞 200자:\n{docs[0].page_content[:200]}")
````

## 2. DirectoryLoader - 폴더 내 여러 파일 로딩

````python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader(
    path="data/",
    glob="*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding": "utf-8"}
)

docs = loader.load()

print(f"총 문서 수: {len(docs)}")
for doc in docs:
    print(f"  - {doc.metadata['source']} ({len(doc.page_content)} 글자)")
````

## 3. WebBaseLoader - 웹 페이지 로딩

````python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://python.langchain.com/docs/introduction/")
docs = loader.load()

print(f"문서 수: {len(docs)}")
print(f"내용 앞 300자:\n{docs[0].page_content[:300]}")
````

## 4. CSVLoader - CSV 파일 로딩

먼저 샘플 CSV를 생성합니다.

````python
import csv

# 샘플 CSV 생성
with open("data/sample.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["기술", "설명"])
    writer.writerow(["LangChain", "LLM 애플리케이션 개발 프레임워크"])
    writer.writerow(["LangGraph", "그래프 기반 멀티 에이전트 워크플로우"])
    writer.writerow(["RAG", "검색 증강 생성 기술"])
    writer.writerow(["FAISS", "Facebook의 벡터 유사도 검색 라이브러리"])

print("CSV 파일 생성 완료")
````

````python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader("data/sample.csv", encoding="utf-8")
docs = loader.load()

print(f"문서 수: {len(docs)}")
for doc in docs:
    print("---")
    print(doc.page_content)
````

## 5. Document 객체 직접 생성

````python
from langchain_core.documents import Document

# 직접 Document 생성
docs = [
    Document(
        page_content="RAG는 검색과 생성을 결합한 기술입니다.",
        metadata={"source": "manual", "topic": "RAG"}
    ),
    Document(
        page_content="LangChain은 LLM 애플리케이션 개발 프레임워크입니다.",
        metadata={"source": "manual", "topic": "LangChain"}
    )
]

for doc in docs:
    print("내용:", doc.page_content)
    print("메타데이터:", doc.metadata)
    print()
````

## 정리

| 로더 | 대상 |
|---|---|
| TextLoader | .txt 파일 |
| PyPDFLoader | .pdf 파일 |
| CSVLoader | .csv 파일 |
| WebBaseLoader | 웹 페이지 URL |
| DirectoryLoader | 폴더 내 다수 파일 |
